```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Color'] = df['Color'].fillna(df['Color'].mode()[0])
```

```python
df['Payment Method'] = df['Payment Method'].fillna(df['Payment Method'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
    
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Young Adult','Adult','Middle aged','Senior']
df['age_group'] =pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(15)
```

```python
df['age'] = df['age'].astype(int)
```

```python
#create purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly':7,
    'Monthly':30,
    'Quaterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months:':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
    
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(15)
```

```python
df['purchase_frequency_days']=df.groupby('category')['purchase_frequency_days'].transform(lambda x: x.fillna(x.median()))
```

```python
df['purchase_frequency_days'] = df['purchase_frequency_days'].astype(int)
```
